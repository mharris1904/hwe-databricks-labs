## Question 7: Categorize Reviews by Sentiment

Create a query that categorizes reviews into sentiment groups based on their star rating:
- **'Positive'** for ratings of 4 or 5
- **'Neutral'** for rating of 3
- **'Negative'** for ratings of 1 or 2

Show the first 10 reviews with their `star_rating` and a new column called `sentiment`.

Then count how many reviews are in each sentiment category.

```sql
-- TODO: Select star_rating and create a sentiment column
SELECT
   *
   , CASE WHEN CAST(star_rating as int) in (4,5) THEN 'Positive'
    WHEN CAST(star_rating as int) in (3) THEN 'Neutral'
    WHEN CAST(star_rating as int) in (1,2) THEN 'Negative'
    END
FROM
  reviews
LIMIT 10
```

```sql
-- TODO: Count reviews by sentiment category
SELECT
  CASE WHEN CAST(star_rating as int) in (4,5) THEN 'Positive'
    WHEN CAST(star_rating as int) in (3) THEN 'Neutral'
    WHEN CAST(star_rating as int) in (1,2) THEN 'Negative'
  END
  , COUNT(*)
FROM
  reviews
GROUP BY ALL
```
