# Create Bronze Schema and Tables

## Setup: Set Your Catalog

**IMPORTANT:** Before running any queries, set your assigned catalog name.


```sql
USE CATALOG redhatmatt;
```

A script to create the bronze layer from scratch. We need to build:

* The `bronze` schema (if it doesn't already exist)
* 5 tables (if they don't already exist):
    * `categories`
    * `stores`
    * `books`
    * `online_orders`
    * `instore_orders`

```sql
-- Create the bronze schema
CREATE SCHEMA IF NOT EXISTS bronze;

```

```sql
-- Create the bronze.categories table
CREATE TABLE IF NOT EXISTS bronze.categories
(
  category_id STRING COMMENT 'Natural key for category'
  , category_name STRING
  , parent_category_key STRING COMMENT 'Empty string for top-level'
  , ingestion_timestamp TIMESTAMP COMMENT 'Added on load'
  , source_filename STRING COMMENT 'Added on load'
  , CONSTRAINT pk_categories PRIMARY KEY (category_id)
) USING DELTA;

```

```sql
-- Create the bronze.stores table
CREATE TABLE IF NOT EXISTS bronze.stores
(
  store_nbr STRING COMMENT 'Natural key'
  , name STRING
  , address STRING
  , city STRING
  , state STRING
  , zip STRING
  , ingestion_timestamp TIMESTAMP COMMENT 'Added on load'
  , source_filename STRING COMMENT 'Added on load'
  , CONSTRAINT pk_stores PRIMARY KEY (store_nbr)
) USING DELTA;
```

```sql
-- Create the bronze.books table
CREATE TABLE IF NOT EXISTS bronze.books
(
  isbn STRING COMMENT 'Natural key'
  , title STRING
  , author STRING
  , category_id STRING COMMENT 'FK to categories (leaf only)'
  , ingestion_timestamp TIMESTAMP COMMENT 'Added on load'
  , source_filename STRING COMMENT 'Added on load'
  , CONSTRAINT pk_books PRIMARY KEY (isbn)
) USING DELTA;
```

```sql
-- Create the bronze.online_orders table
CREATE TABLE IF NOT EXISTS bronze.online_orders
(
  order_id STRING COMMENT 'Natural key'
  , order_timestamp TIMESTAMP
  , customer_email STRING
  , customer_name STRING
  , customer_address STRING COMMENT 'Street address'
  , customer_city STRING
  , customer_state STRING
  , customer_zip STRING
  , items STRING COMMENT 'JSON Array'
  , payment_method STRING
  , total_amount DECIMAL(10,2)
  , ingestion_timestamp TIMESTAMP COMMENT 'Added on load'
  , source_filename STRING COMMENT 'Added on load'
  , CONSTRAINT pk_online_orders PRIMARY KEY (order_id)
) USING DELTA;
```

```sql
-- Create the bronze.instore_orders table
CREATE TABLE IF NOT EXISTS bronze.instore_orders
(
  order_id STRING COMMENT 'Natural key'
  , transaction_timestamp TIMESTAMP
  , store_nbr STRING
  , customer_email STRING COMMENT 'Nullable'
  , items STRING COMMENT 'JSON array'
  , payment_method STRING
  , total_amount DECIMAL(10,2)
  , cashier_name STRING
  , ingestion_timestamp TIMESTAMP COMMENT 'Added on load'
  , source_filename STRING COMMENT 'Added on load'
  , CONSTRAINT pk_instore_orders PRIMARY KEY (order_id)
) USING DELTA;
```
